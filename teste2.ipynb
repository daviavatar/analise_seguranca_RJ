```python
# lista de ocorrencias relacionadas a veiculos
ocorrencias_veiculo = ['furto_veiculos', 'roubo_veiculo']

#juntando as 3 colunas necessárias
df_veiculos= df.groupby('cisp')[ocorrencias_veiculo].sum().reset_index()
df_veiculos
```
